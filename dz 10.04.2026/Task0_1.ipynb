```python
import ipywidgets as widgets
from IPython.display import display

s1 = widgets.SelectionRangeSlider(options=list(range(101)), index=(0, 5), description='N1')
op = widgets.Dropdown(options=['+', '-', '*', '/'], description='Op')
s2 = widgets.SelectionRangeSlider(options=list(range(101)), index=(0, 5), description='N2')

def calc(a, o, b):
    v1, v2 = a[0], b[0]
    if o == '+': res = v1 + v2
    elif o == '-': res = v1 - v2
    elif o == '*': res = v1 * v2
    elif o == '/': res = v1 / v2 if v2 != 0 else 'Error'

    print(f"| {v1} | {v2} | {res} |")

widgets.interact(calc, a=s1, o=op, b=s2)
```

```python
import ipywidgets as widgets

banwords = ['плохо', 'спам', 'реклама']
inp = widgets.Textarea(description='Text:')

def filter_text(txt):
    res = txt
    for word in banwords:
        res = res.replace(word, '***')
    print(res)

widgets.interact(filter_text, txt=inp)
```
