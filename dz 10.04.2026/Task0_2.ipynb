```python
import ipywidgets as widgets
import seaborn as sns
import pandas as pd

df = sns.load_dataset('iris')

col_choice = widgets.Dropdown(options=df.columns, description='Колонка:')
row_slider = widgets.IntSlider(min=1, max=20, value=5, description='Строк:')

def explore(col, n):
    print(f"Первые {n} строк колонки {col}:")
    print(df[col].head(n))
    print("\nСтатистика по колонке:")
    print(df[col].describe())

widgets.interact(explore, col=col_choice, n=row_slider)
```

```python
import ipywidgets as widgets
import seaborn as sns

df = sns.load_dataset('iris')
val_slider = widgets.FloatSlider(min=1.0, max=7.0, step=0.1, value=1.0, description='Min Length:')

def show_data(min_val):
    filtered_df = df[df['petal_length'] > min_val]
    display(filtered_df.head(10))

widgets.interact(show_data, min_val=val_slider)
```
