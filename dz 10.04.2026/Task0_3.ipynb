```python
import pandas as pd
import ipywidgets as widgets

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/adult-all.csv"
columns = ["age", "workclass", "fnlwgt", "education", "education_num", "marital_status",
           "occupation", "relationship", "race", "sex", "capital_gain",
           "capital_loss", "hours_per_week", "native_country", "income"]

df = pd.read_csv(url, names=columns, na_values="?", skipinitialspace=True)

def check_hypotheses(hypothesis):
    print(f"--- {hypothesis} ---\n")

    if hypothesis == '1. Образование и часы':
        res = df.groupby('education')['hours_per_week'].mean().sort_values()
        print(res)
        print("\nГипотеза опровергнута")

    elif hypothesis == '2. Пол и часы':
        res = df.groupby('sex')['hours_per_week'].mean()
        print(res)
        print("\nГипотеза доказана")

    elif hypothesis == '3. Доход и капитал':
        res = df.groupby('income')['capital_gain'].apply(lambda x: (x > 0).mean() * 100)
        print("Процент людей с доходом от капитала (%):")
        print(res)
        print("\nГипотеза доказана")

hypo_dropdown = widgets.Dropdown(
    options=['1. Образование и часы', '2. Пол и часы', '3. Доход и капитал'],
    description='Гипотеза:'
)

widgets.interact(check_hypotheses, hypothesis=hypo_dropdown)
```
